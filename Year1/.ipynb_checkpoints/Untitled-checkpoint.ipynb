```python
apollo = cv2.imread('./OpenCV/03_Image_Annotation/Apollo_11_Launch.jpg', cv2.IMREAD_COLOR)
plt.imshow(apollo[:,:,::-1])
#img = cv2.line(img, pt1, pt2, color, thickness, linetype, shift)
```
